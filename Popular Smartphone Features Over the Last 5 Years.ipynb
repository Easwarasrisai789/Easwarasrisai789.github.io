```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('smartphone_features.csv')

# Print the available columns to check for the correct column names
print(data.columns)

# Update column names to the correct ones if necessary.
# Assuming the actual columns are named 'brand' and 'popularity'
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='             Brand', y='Popularity Score', errorbar=None, palette='viridis')
plt.title('Popularity Score by Brand')
plt.xlabel('Brand')
plt.ylabel('Popularity Score')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Removing the leading space from ' Year'
sns.lineplot(data=data, x='               Year', y='Popularity Score', hue='             Brand', marker='o', palette='tab10')
plt.title('Popularity Score Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Popularity Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Popularity Score', y='                   Feature Name', hue='              Category', style='             Brand', s=100, palette='coolwarm')
plt.title('Popularity Score vs. Features')
plt.xlabel('Popularity Score')
plt.ylabel('Feature Name')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Popularity Score'], bins=20, kde=True, color='blue')
plt.title('Distribution of Popularity Scores')
plt.xlabel('Popularity Score')
plt.ylabel('Frequency')
plt.show()
```
